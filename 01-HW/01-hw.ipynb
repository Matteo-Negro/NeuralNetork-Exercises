```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.uniform(-2, 2, 1000)
y = np.random.uniform(-2, 2, 1000)
```

```python
plt.scatter(x, y)
```

```python
W1 = np.array([[ 1, -1], [-1, -1], [-1,  0]])

b1 = np.array([[1, 1, 0]]).T

W2 = np.array([[1, 1, 1]])

b2 = np.array([[-1.5]]).T
```

```python
x = np.expand_dims(x, axis=1)
y = np.expand_dims(y, axis=1)
X = np.hstack((x, y)).T
```

```python
X.shape, W1.shape, b1.shape, (np.heaviside(W1 @ X + b1, np.zeros(1000))).shape
```

```python
Z = np.heaviside(W2 @ (np.heaviside(W1 @ X + b1, np.zeros(1000))) + b2, np.zeros(1000))
Z.shape
```

```python
for i in range(X.shape[1]):
    if Z[0][i] == 0:
        c = 'b'
    else:
        c = 'r'
    plt.scatter(X[0][i], X[1][i], color=c)
```
